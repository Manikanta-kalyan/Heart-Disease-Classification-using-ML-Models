## Decision Tree Classification For Heart Diseases DataSet

## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('D:/archive1/heart_Disease.csv')
dataset.head()
```

## Correlation between features

```python
dataset.corr()
```

## Dataset Description

```python
dataset.describe()
```

## Input and output feature selection

```python
X=dataset.iloc[:, 0:13].values
y=dataset.iloc[:, 13].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Fitting Decision Tree Classification to the Training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

## Model Accuracy and Scores

```python
print ('Classifier',classifier.score(X_test, y_test))
print('roc_auc: ', roc_auc_score(y_test, y_pred))
print('precision: ', precision_score(y_test, y_pred))
print('recall: ', recall_score(y_test, y_pred))
print('f1: ', f1_score(y_test, y_pred))
```
